## Шаг 4: Группировка и агрегация данных

Используйте `df.groupby()` для группировки данных по определенному столбцу.

Примените агрегирующие функции, такие как `.mean(), .sum(), .count()` к сгруппированным данным.

```python
season_avg_cnt = df.groupby('season')['cnt'].mean()
print(season_avg_cnt , '\n\n')

holiday_casual_sum = df.groupby('holiday')['casual'].sum()
print(holiday_casual_sum, '\n\n')

weekday_count = df.groupby('weekday')['season'].count()
print(weekday_count, '\n\n')
```

## Шаг 6: Заполнение и обработка данных

Если есть пропущенные значения, используйте .fillna() или .dropna() для их обработки.

Когда в данных есть пропущенные значения, вы можете использовать метод .fillna() для заполнения их определенным значением или стратегией, либо метод .dropna() для удаления строк или столбцов с пропущенными данными. Давайте рассмотрим примеры использования обоих методов:

Предположим, у вас есть DataFrame df, и в столбце "age" есть пропущенные значения, которые вы хотите обработать.

Использование .fillna():

Заполнение пропущенных значений средним значением столбца:
python
Copy code
df['age'].fillna(df['age'].mean(), inplace=True)
Это заполнит пропущенные значения в столбце "age" средним значением этого столбца.

Использование .dropna():

Удаление строк с пропущенными значениями:
python
Copy code
df.dropna(subset=['age'], inplace=True)
Это удалит все строки, в которых столбец "age" содержит пропущенные значения.

Удаление столбца с пропущенными значениями:
python
Copy code
df.dropna(axis=1, inplace=True)
Это удалит столбец "age" целиком, если он содержит хотя бы одно пропущенное значение.

Обратите внимание, что inplace=True используется для применения изменений к исходному DataFrame df. Если вы не используете inplace=True, то изменения будут создавать новый DataFrame.

Выбор между .fillna() и .dropna() зависит от вашей конкретной задачи и данных. .fillna() полезен, когда вы хотите сохранить пропущенные значения, но заменить их на какие-то значения, в то время как .dropna() полезен, когда вы предпочитаете удалить пропущенные данные для проведения анализа.






# Шаг 7: Создание новых столбцов

Создайте новые столбцы, которые могут быть полезными для вашего анализа.

создать новый столбец с текстовыми значениями, чтобы сделать данные более интерпретируемыми:

```python
season_mapping = {2: 'весна', 3: 'лето', 4: 'осень', 1: 'зима'}
df['season_name'] = df['season'].map(season_mapping)
df
```

Категоризация температуры: Если у вас есть столбцы temp и atemp, которые представляют температуру и ощущаемую температуру, вы можете создать категориальный столбец, например, "холодно", "тепло" и "жарко", на основе диапазонов температур:

```python
def categorize_temperature(temp):
    if temp < 0.2:
        return 'холодно'
    elif 0.2 <= temp < 0.3:
        return 'тепло'
    else:
        return 'жарко'

df['temperature_category'] = df['temp'].apply(categorize_temperature)
df.head()
```


Метод `.apply()` в библиотеке Pandas используется для применения пользовательской функции к элементам DataFrame или Series. Этот метод позволяет вам выполнять более сложные операции на данных, чем стандартные методы.

# Шаг 3: Работа с множеством графиков

### Создание множества графиков:

Создайте несколько областей рисования (subplots) для одного графика.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

Отображение множества графиков:

Настройте подграфики и отобразите несколько графиков на одной фигуре.

```python
# Пример данных
x1 = [1, 2, 3, 4, 5]
y1 = [10, 14, 18, 24, 30]

x2 = [1, 2, 3, 4, 5]
y2 = [8, 12, 15, 20, 25]

x3 = [1, 2, 3, 4, 5]
y3 = [15, 10, 13, 17, 20]

# Создание фигуры с подграфиками 2x2
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# Построение графика на верхнем левом подграфике
axes[0, 0].plot(x1, y1, label='График 1')
axes[0, 0].set_xlabel('X1')
axes[0, 0].set_ylabel('Y1')
axes[0, 0].legend()

# Создание scatter plot на верхнем правом подграфике
axes[0, 1].scatter(x2, y2, label='График 2')
axes[0, 1].set_xlabel('X2')
axes[0, 1].set_ylabel('Y2')
axes[0, 1].legend()

# Создание bar plot на нижнем левом подграфике
axes[1, 0].bar(x3, y3, label='График 3')
axes[1, 0].set_xlabel('X3')
axes[1, 0].set_ylabel('Y3')
axes[1, 0].legend()

# Создание bar plot на нижнем правом подграфике
axes[1, 1].bar(x2, y2, label='График 4')
axes[1, 1].set_xlabel('X2')
axes[1, 1].set_ylabel('Y2')
axes[1, 1].legend()

# Отображение графиков
plt.show()
```
